```python
import numpy as np, pandas as pd, unicodedata, pickle, re
```

```python
txtfilename = "quotes.txt"
txtfile = open(txtfilename)
```

```python
r = re.compile("(.*?)(http(s?)://.+)?(\s\w{3}\s\d{2},\s\d{4})$")
def clean(s):
    m = r.match(s)
    return m.group(1)
    
```

```python
clean('"qsdfmlqsmdflkj qsmdlkfj mqsdlfkj " - Qqsdf qsdf ht qsdf Dec 15, 2015')
```

```python
def char2idx(c):
    return ord(c)
def toascii(s):
    return unicodedata.normalize("NFKD", s).encode("ascii", "ignore")
def tostring(lc):
    return "".join(map(chr, lc))
```

```python
with open("quotes.clean.txt","w") as outf:
    for line in txtfile:
        try: 
            outf.write(clean(line)+"\n")
        except AttributeError:
            print line
```

```python
txtfile = open("quotes.clean.txt")

datamat = []
for line in txtfile:
    idxs = map(char2idx, toascii(line.decode("utf8")))
    datamat.append(idxs)
```

```python
df = pd.DataFrame(datamat)
```

```python
data = df.fillna(0).values
```

```python
savefile = "quotes.np.pkl"
pickle.dump(data, open(savefile, "w"))
```

```python
sdata = pickle.load(open(savefile))
```
